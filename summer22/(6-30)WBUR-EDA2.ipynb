## Calculate duration of cases

We will only check rows where case_status is Disposed and Closed

```python
Case_df['file_date'] = pd.to_datetime(Case_df['file_date'],  errors='coerce')
Case_df['status_date'] = pd.to_datetime(Case_df['status_date'],  errors='coerce')
Case_df['Days'] = (Case_df['status_date'] - Case_df['file_date']).dt.days
Case_df = Case_df.loc[Case_df['Days'] > 0]
Case_df.head(5)
```

### Find judges name based on judge IDs

```python
Case_df = Case_df.loc[ (Case_df['judge_1_id'] != 0) | (Case_df['judge_2_id'] != 0) | (Case_df['judge_3_id'] != 0) | (Case_df['judge_4_id'] != 0) | (Case_df['judge_5_id'] != 0)]
Case_df

```

## Which judges have long average time?

```python
# replace judge with names first by matching id
Case_df['judge_1_id'] = Case_df['judge_1_id'].map(judges_id.set_index('ID')['post_title'])
Case_df['judge_2_id'] = Case_df['judge_2_id'].map(judges_id.set_index('ID')['post_title'])
Case_df['judge_3_id'] = Case_df['judge_3_id'].map(judges_id.set_index('ID')['post_title'])
Case_df['judge_4_id'] = Case_df['judge_4_id'].map(judges_id.set_index('ID')['post_title'])
Case_df['judge_5_id'] = Case_df['judge_5_id'].map(judges_id.set_index('ID')['post_title'])
```

```python
df = pd.read_csv('Mass_Cases(More).csv')
df.head(2)
```

```python
result0 = Case_df.groupby(['judge_1_id']).agg({'Days': ['mean', 'min', 'max', 'std']})
result0
```

## Finding Cases Closed

```python
Closed_df = Case_df.loc[Case_df['case_status'] == 'Closed']
Disposed_df = Case_df.loc[Case_df['case_status'] == 'Disposed (Statistical Purposes)']
print('Average time for cases to be closed is', math.floor(Closed_df['Days'].mean()), 'Days')
print('Average time for cases to be Disposed (Statistical Purposes) is', math.floor(Disposed_df['Days'].mean()), 'Days')
```
