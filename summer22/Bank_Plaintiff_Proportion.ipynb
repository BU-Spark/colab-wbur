```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/drive/My Drive/SPARK WBUR SUMMER 2022/WBUR data/Data description/Data/updated_case_meta.csv')
```

```python
df1
```

```python
df1 = df1[['post_id','case_status','status_date']]
# Filter anamoly
df1 = df1[df1['status_date'] > 13140820]
df1 = df1[df1['case_status'].isin(['Disposed (Statistical Purposes)','Closed','Closed (Transferred)','Suspended'])]
df1['status_date'] = pd.to_datetime(df1.status_date, format="%Y%m%d")
df1
```

```python
df1[df1['status_date']<datetime(2001,1,1,0,0)]
# Only fewer than 100 cases before 2001
```

```python
df2 = pd.read_csv('/content/drive/My Drive/SPARK WBUR SUMMER 2022/WBUR data/Data description/Data/updated_plain_def.csv')
df2
```

```python
df2vs = pd.DataFrame(df2['text'].str.split('vs.', 1).tolist(),
                        columns = ['plaintiff', 'defendant'])
df2vs['post_id'] = df2['post_id']
df2vs = df2vs[['post_id', 'plaintiff']]
df2vs
```

```python
df2v = pd.DataFrame(df2vs['plaintiff'].str.split(' v. ', 1).tolist(),
                        columns = ['plaintiff', 'defendant'])
df2v['post_id'] = df2vs['post_id']
df2v = df2v[['post_id', 'plaintiff']]
df2v
```

```python
df2v['plaintiff'] = df2v['plaintiff'].str.upper()
df2 = df2v
df2
```

```python
df2["BANK"] = 0
for i in range(len(df2["BANK"])):
    if "BANK" in df2['plaintiff'][i]:
        df2["BANK"][i] = 1
#df3["BANK"] = [0 if 'DISCOVER BANK' in df3['plaintiff'] else 1 for x in df3["BANK"]]
#df3["BANK"] = [0 if df3['post_id'] == 5464 else 1 for x in range(len(df3["BANK"]))]
df2
```

```python
print(df2["BANK"].value_counts())
df2["BANK"].value_counts(normalize=True)
```

```python
df12 = pd.merge(df1, df2, on = "post_id", how="inner")
df12
```

```python
df12['status_date'].max()
```

# Group BY Year and Count Proportion of Bank as Plaintiff

```python
#df12.groupby(pd.Grouper(key='status_date', freq='Y', sort = True))['BANK'].value_counts(normalize = True).plot(kind='bar')
df12.groupby(pd.Grouper(key='status_date', freq='Y', sort = True))['BANK'].value_counts()
```

```python
# Before 2004, only a small amount of cases happen, so the proportion that year is not meaningful in lack of samples.
df12 = df12[df12['status_date']>datetime(2003,12,31,0,0)]
df12
```

```python
print(df12.groupby(pd.Grouper(key='status_date', freq='Y', sort = True))['BANK'].value_counts(normalize = True).unstack(fill_value=0).loc[:,1].tolist())
print(df12.groupby(pd.Grouper(key='status_date', freq='Y', sort = True))['BANK'].value_counts())
print(df12.groupby(pd.Grouper(key='status_date', freq='Y', sort = True))['BANK'].value_counts(normalize = True))
```
