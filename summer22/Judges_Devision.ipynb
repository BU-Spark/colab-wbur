# Judge Average Days in Past Closed Cases

Some cases have multiple judges. When counting judge average days and frequency, we need to consider all judges in one case, and add all cases up.

```python
df_rep = []
for i in range(len(data)):
    # up to 5 judges
    if pd.isna(data.loc[i, "judge_5_id"]) == False:
        row_rep = data.loc[i,:]
        row_rep["judge_1_id"] = row_rep["judge_5_id"]
        row_rep["judge_2_id","judge_3_id","judge_4_id","judge_5_id"] = np.NaN
        df_rep.append(row_rep)
    if pd.isna(data.loc[i, "judge_4_id"]) == False:
        row_rep = data.loc[i,:]
        row_rep["judge_1_id"] = row_rep["judge_4_id"]
        row_rep["judge_2_id","judge_3_id","judge_4_id","judge_5_id"] = np.NaN
        df_rep.append(row_rep)
    if pd.isna(data.loc[i, "judge_3_id"]) == False:
        row_rep = data.loc[i,:]
        row_rep["judge_1_id"] = row_rep["judge_3_id"]
        row_rep["judge_2_id","judge_3_id","judge_4_id","judge_5_id"] = np.NaN
        df_rep.append(row_rep)
    if pd.isna(data.loc[i, "judge_2_id"]) == False:
        row_rep = data.loc[i,:]
        row_rep["judge_1_id"] = row_rep["judge_2_id"]
        row_rep["judge_2_id","judge_3_id","judge_4_id","judge_5_id"] = np.NaN
        df_rep.append(row_rep)

data_5judges = data.append(df_rep,ignore_index=True)
data_5judges.shape
```

```python
data_5judges.head()
```

```python
# We only need to count data fall in the following statuses
# For cases that is still active, we don't know how long it lasts
options = ['Disposed (Statistical Purposes)', 'Closed', 'Closed (Transferred)', 'Suspended'] 
    
data_5judges_disposedclosed = data_5judges[data_5judges['case_status_x'].isin(options)] 
data_5judges_disposedclosed
```

```python
data_5judges_sum = data_5judges_disposedclosed.groupby('judge_1_id').sum()
data_5judges_sum = data_5judges_sum.drop(columns=['Unnamed: 0', 'post_id'])
data_5judges_sum
```

```python
# Average Days
data_5judges_mean = data_5judges_disposedclosed.groupby('judge_1_id').mean()
data_5judges_mean = data_5judges_mean.drop(columns=['Unnamed: 0', 'post_id'])
data_5judges_mean
```

```python
mean = data_5judges_mean["Days"]
data_5judges_sum['mean'] = mean
data_5judges_sum
```

```python
data_5judges_sum.to_csv("Judges_Duration.csv")
```
