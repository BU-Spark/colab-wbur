```python
debt = pd.read_excel('Debt.xlsx')
```

```python
l = debt['Unnamed: 1'].unique()
```

```python
n = []
t = []
status = []
year = []
```

```python
for i in range(1,len(l)):
    s = [x.replace(',','') for x in l[i].split()]
    v = '%'
    for w in s:
        v = v + w + '%'
    q = query
    q = q + '"' + v + '"'
    with mydb.cursor() as cursor:
        cursor.execute(q)
        result = cursor.fetchall()
    if len(result) != 0:
        for res in result:
            n.append(res[0])
            t.append(res[2])
            status.append(res[3])
            year.append(res[-1])
        
```

```python
d = {'name':n,'type':t,'status':status,'year':year}
```

```python
df = pd.DataFrame(data=d)
df
```
