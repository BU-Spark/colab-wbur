```python
def count_cases_in_2020(filepath):
    counter = {}
    with open(filepath, "r") as f:
        reader = csv.reader(f)
        header = next(reader)
        for line in reader:
            #collector_cases += [line]

            if str(line[-2])[:4] == "2020":
                start_date = pd.to_datetime(str(line[-2]))
                if start_date in counter:
                    counter[start_date] += 1
                else:
                    counter[start_date] = 1

    temp = [(a, counter[a]) for a in counter.keys()]
    temp.sort(key=lambda a : a[0])

    x = [a[0] for a in temp]
    y = [a[1] for a in temp]

    # plot
    plt.plot(x,y)
    # beautify the x-labels
    plt.gcf().autofmt_xdate()

    plt.title(filepath)

    
    plt.savefig(os.path.basename(filepath)+".png")
    plt.show()
#collector_cases
```

```python
def count_cases_in_2020_by_month(filepath):
    counter = {}
    with open(filepath, "r") as f:
        reader = csv.reader(f)
        header = next(reader)
        
        for i in ["2018", "2019", "2020"]:
            for j in ["01","02","03","04","05","06","07","08","09","10","11","12"]:
                counter[i+str(j)] = 0
        
        for line in reader:
            #collector_cases += [line]

            if str(line[-2])[:4] in ["2018","2019","2020"]:
                start_date = str(line[-2])[:6]
                if start_date in counter:
                    counter[start_date] += 1
                else:
                    counter[start_date] = 1

    temp = [(a, counter[a]) for a in counter.keys()]
    temp.sort(key=lambda a : a[0])

    x = [a[0] for a in temp]
    y = [a[1] for a in temp]

    # plot
    plt.bar(x,y)
    # beautify the x-labels
    plt.gcf().autofmt_xdate()
    
    plt.title(filepath)
    
    plt.rcParams["figure.figsize"] = (18,10)

    
    
    plt.savefig(os.path.basename(filepath)+".png")
    plt.show()
```

```python
def count_cases_in_2020_by_week(filepath):
    counter = {}
    with open(filepath, "r") as f:
        reader = csv.reader(f)
        header = next(reader)
        
        for j in range(55):
            counter[j] = 0
        
        for line in reader:
            #collector_cases += [line]

            if str(line[-2])[:4] in ["2020"]:
                start_date = pd.to_datetime(str(line[-2]))
                week_num = start_date.isocalendar().week
                if week_num in counter:
                    counter[week_num] += 1
                else:
                    counter[week_num] = 1

    temp = [(a, counter[a]) for a in counter.keys()]
    temp.sort(key=lambda a : a[0])

    x = [a[0] for a in temp]
    y = [a[1] for a in temp]

    # plot
    plt.bar(x,y)
    # beautify the x-labels
    plt.gcf().autofmt_xdate()
    
    plt.title(filepath)
    
    plt.rcParams["figure.figsize"] = (18,10)

    
    
    plt.savefig(os.path.basename(filepath)+".png")
    plt.show()
```

```python
for filepath in top_debt_collectors:
    count_cases_in_2020_by_week(filepath)
```
